```python
import numpy as np
import matplotlib.pyplot as plt
def graficacion(x):
  x=np.linspace(0,6*np.pi,150)
  sen=np.sin(x)

  plt.subplot(1, 2, 1)
  plt.title("senos")
  plt.xlabel("x")
  plt.ylabel("y")

  plt.plot(x,sen)
  plt.grid(color = 'magenta', linestyle = '--', linewidth = 0.5)

  x=np.linspace(0,6*np.pi,150)
  cos=np.cos(x)

  plt.subplot(1, 2, 2)
  plt.title("cosenos")
  plt.xlabel("x")
  plt.ylabel("y")

  plt.plot(x,cos,'d-b')
  plt.grid(color = 'cyan', linestyle = '-.', linewidth = 0.5)
  plt.suptitle("Senos y Cosenos")
  plt.show()
  return
graficacion(x)
```
