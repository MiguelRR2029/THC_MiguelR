1.- Con base en la definición de número primo, escriba una función que reciba un número entero e imprima 'True' si es primo o de lo contrario imprima 'False'. Tenga en cuenta que si el número es menor o igual a 1, se retorna como resultado "Error en el Número" y debe volver a preguntar hasta que la entrada sea valida.

```python
#Pedimos un número entero
u=int(input("Este número es primo: "))
#Creamos una lista vacia
div=[]
#Creamos una función para ver si el residuo de dos numeros es 0 agregarlos a la lista vacía
for i in range(1, u + 1):
    if u % i == 0:
        div.append(i)
#Si la lista tiene dos elementos el número es primo
if len(div) == 2:
  print(u, "True")
#Si la lista tiene mas de dos elementos no es primo
elif len(div) > 2:
  print(u, "False")
#Imprimo los divisores
print(div)
```

2.- Sin utilizar librerías adicionales cree un arreglo de numéros enteros (entre 0 y 150) y calcule a través de funciones estándar o lambda lo siguiente:

*   El promedio

*   La desviación estándar
*   Visualice los datos en un histograma.
*   La suma de los números pares e impares.
*   Los múltiplos de 2, 3 y 5.

```python
#Primero creamos una lista de 5 números aleatorios entre 1 y cien
#definimos una funcion para crear la lista
def listaAleatorios(n):
#creamos una lista vacía
    l=[]
#Pedimos que para el número de elemtos dados se de una lista alaeatoria
    for i in range(n):
        l.insert(i, random.randrange(0, 151))
    return l
n=5
A = listaAleatorios(n)
print("Dda una lista aleatoria",A)
#Promedio
s=0
#Creamos un bucle para la suma de todos los numero del 1 al 150
for n in A:
  s=s+n
#dividimos el resultado entre 5 para sacar el promedio
p=s/5
print("El promedio de 5 números aleatorioas entre el 1 y 150 son: ",p)

#Varianza cabe aclarar que la fórmula en la que nos basamos es la de población
V=0
#Creamos bucle que sume el cuadrado de la suma de cada número mas su promedio
for n in A:
 V=(n-p)**2+V
#Dividemos el resultado entre el número de números ocupados en este caso 5
Va=V/5
print("La varianza de 5 números aleatorioas entre el 1 y 150 son: ",Va)

#Desviación estándar cabe aclarar que la fórmula en la que nos basamos es la de población
#Sacamos la raiz cuadrada de la varianza
Y=['Promedio','Varianza','desviación etándar']
DE=Va**(1/2)
print("La desviación estándar de 5 números aleatorioas entre el 1 y 150 es: ",DE)

#Histograma
#Damos lo que vamos a graficar
g=[p,Va,DE]
#Hacemos una gráfica de barras
plt.bar(Y,g,color='cyan',alpha=0.5)
plt.grid()
plt.show()

#La suma de los números pares e impares.
#Creamos dos listas vacias
par=[]
impar=[]
#Creamos un bucle para la suma de todos los numero del 1 al 150
for n in range(1,151):
  # Pedimos el reciduo de la operación n/2
  i=n%2
  #Creamos un if donde el reciduo sea igual a 0
  if i==0:
    #Agregamos el elemento que cumple a la lista vacia par
    par.append(n)
    s=0
    #Creamos bucle que sume los pares
    for n in par:
      s=n+s
    #Creamos un elif donde el reciduo sea diferente a 0
  elif not i==0:
      #Agregamos el elemento que cumple a la lista vacia impar
    impar.append(n)
    si=0
      #Creamos bucle que sume los impares
    for n in impar:
      si=n+si
print("La suma de los pares es:",s)
print("La suma de los impares es:",si)

#Los múltiplos de 2, 3 y 5.
#Creamos una lista vacía para cada multipplo
dos=[]
tres=[]
cinco=[]
#Pedimos que se rellen las listas vacías con los numeros que cumpla que el reciduo sea 0 para cada número dividido entre los numerosdel 1 al 150
for n in range(1,151):
  d=2*n
  if d < 151:
    dos.append(d)
for n in range(1,151):
  t=3*n
  if t < 151:
    tres.append(t)
for n in range(1,151):
  c=5*n
  if c < 151:
    cinco.append(c)
print("Los multiplos de 2 entre el 1 y 150 son ", dos)
print("Los multiplos de 3 entre el 1 y 150 son ",tres)
print("Los multiplos de 5 entre el 1 y 150 son ",cinco)
```

3.- En la clase de ecuaciones diferenciales se pide a los estudiantes encontrar la solución del problema de valor inicial:

\begin{equation}
y´=-y+x \hspace{3cm} y(-1)=5.
\end{equation}


Los estudiantes llegaron a la solución general cuya ecuación es:

\begin{equation}
y=(x-1)+C e^{-x}
\end{equation}

Apoye a sus compañeros y grafique en python la familia de curvas para la solución de la ecuación diferencial y en un sólo gráfico muestre de diferente color las curvas para los siguientes tres casos:

*   para valores de C > 0
*   cuando C = 0
*   para valores de C < 0

Señale en el gráfico las curvas que corresponden a cada uno de los tres casos.

```python
#Le damos a C un valor mayor a 0
C=0.001
#Definimos su función
def f(x):
  return (x-1)+(C*np.e**-x)
#En el eje x va a ir de -10 a 10
x=np.linspace(-10,10,1000)
#Graficamos la gráfica
plt.plot(x, f(x),'-r')
#Le damos a C un valor igual a 0
C=0
#Definimos su función
def f(x):
  return (x-1)+(C*np.e**-x)
#En el eje x va a ir de -10 a 10
x=np.linspace(-10,10,1000)
#Graficamos la gráfica
plt.plot(x, f(x),'.-b')
#Le damos a C un valor igual a 0
C=-0.001
#Definimos su función
def f(x):
  return (x-1)+(C*np.e**-x)
#En el eje x va a ir de -10 a 10
x=np.linspace(-10,10,1000)
#Graficamos la gráfica
plt.grid()
plt.title('Gráfica $y=(x-1)+C e^{-x}$')
plt.plot(x, f(x),'--c')
A='C>0','0=0','C<0'
plt.legend(A)

plt.show()
```
