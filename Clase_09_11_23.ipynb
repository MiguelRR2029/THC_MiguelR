```python
import pandas as pd
import matplotlib.pyplot as plt
link='https://github.com/seismomat/HerramientasComputacionales/blob/main/Aplicaciones/ebola_outbreak_dataset-master/ebola_outbreak_dataset-master/ebola_2014_2016_clean.csv'
code='?raw=true'
url=link+code
df=pd.read_csv(url,sep=',',index_col=1)
```

```python
SL=df[df['Country']=='Sierra Leone']
SL
```

```python
SL=SL.drop('Country', axis=1)
```

```python
SL.head()
```

```python
SL[15:20]
```

```python
SL.shape
```

```python
SL.isnull().sum()
```

```python
mean_values=SL['No. of confirmed deaths'].mean()
mean_values
```

```python
SL.fillna(mean_values,inplace=True)
#clase09/11
```

```python
SL.isnull().sum()
```

```python
SL['No. of confirmed deaths'].mean()
```

```python
SL['No. of confirmed cases'].mean()
```

```python
Dates=SL.index[0:-1:10]
NCC=SL['No. of confirmed cases'][0:-1:10]
NCD=SL['No. of confirmed deaths'][0:-1:10]
```

```python
fig=plt.figure(figsize=(16,8))
plt.plot(Dates,NCC,'d-.b',ms=4,label='infectados')
plt.bar(Dates,NCC,color='blue',alpha=0.5)
plt.plot(Dates,NCD,'o--c',ms=4,label='decesos')
plt.bar(Dates,NCD,color='cyan',alpha=0.9)
plt.grid()
plt.title('Avanza del ébola en Sierra Leona')
plt.ylabel('Casos confirmados')
plt.xlabel('Periodo de tiempo')
plt.xticks(rotation=90,fontsize=8)
plt.legend()
plt.savefig("Barras_SierraLeona.png",dpi=300,format='png')
plt.show()
#clase09/11
```

```python
PIvsD=100*SL['No. of confirmed deaths']/SL['No. of confirmed cases']
SL['PIvsD']=PIvsD
```

```python
SL.head()
```

```python
recuperados=((100-PIvsD)/100)*SL['No. of confirmed cases']
SL['recuperados']=recuperados
```

```python
SL['PR']=(100-PIvsD)
#clase09/11
```

```python
SL.head()
```

```python
PIvsD_mean=SL['PIvsD'].mean()
Pr_mean=SL['PR'].mean()
porcentajes=[PIvsD_mean,Pr_mean]
etiquetas=['PIvsD','PR']
#clase09/11
```

```python
plt.pie(porcentajes,labels=etiquetas,autopct='%0.1f %%',colors=['red','blue'])
plt.axis("equal")
plt.title("Porcengtaje de recuperados contra decesos")
plt.show()
#clase09/11
```
