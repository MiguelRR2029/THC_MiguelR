```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,6*np.pi,150)
cos=np.cos(x)
sen=np.sin(x)


plt.title("Título-senos y cosenos")
plt.xlabel("x")
plt.ylabel("y")

plt.plot(x,cos,x,sen)
plt.grid(color = 'magenta', linestyle = '--', linewidth = 0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def graficacion(x):
  cos=np.cos(x)
  sen=np.sin(x)
  return

x=np.linspace(0,6*np.pi,150)

graficacion(x)


plt.title("Título-senos y cosenos")
plt.xlabel("x")
plt.ylabel("y")

plt.plot(x,cos,'d-b',x,sen,'*-k')
plt.grid(color = 'cyan', linestyle = '-.', linewidth = 0.5)
plt.show()
```
